```python
import random

print("Halo! Selamat datang di permainan tebak angka.\nKamu punya 7 kesempatan untuk menebak angka. Ayo mulai!")

low = int(input("Masukkan angka batas bawah : "))
high = int(input("Masukkan angka batas atas : "))

print(f"\nKamu punya 7 kesempatan untuk menebak angka antara {low} dan {high}. Ayo mulai!")

num = random.randint(low, high)
ch = 7                        # Total allowed chances
gc = 0                        # Guess counter

while gc < ch:
    gc += 1
    guess = int(input('Ketikkan tebakanmu: '))

    if guess == num:
        print(f'Benar! Angka nya adalah {num}. Kamu menebaknya dalam {gc} percobaan.')
        break

    elif gc >= ch and guess != num:
        print(f'Maaf! Angka yang benar adalah {num}. Coba lagi lain kali.')

    elif guess > num:
        print('Terlalu tinggi! Coba dengan angka yang lebih rendah.')

    elif guess < num:
        print('Terlalu rendah! Coba dengan angka yang lebih tinggi.')
```
